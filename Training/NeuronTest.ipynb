```python
import numpy as np
from neuron import *
import matplotlib.pyplot as plt
```

```python
neuron_param = NeuronParam(t_refractory = 5,
                           V_rest = 0,
                           V_thr = 25,
                           V_min = 0,
                           leak = 0.25)

neuron = SimpleNeuron(neuron_param)
```

```python
T = 50;
dt = 0.125;
time = np.arange(0,T+dt,dt)
S=np.random.randint(0,2,len(time))
```

```python
spike_train = []
for i,t in enumerate(time):
    spike = neuron.run(S[i],t)
    V = neuron.V
    if spike:
        V = 50
    spike_train.append(V)
    
spike_train=np.array(spike_train)
```

```python
plt.figure(figsize=(18,6))
plt.plot(spike_train)
plt.stem(S*20,markerfmt=" ",linefmt='C1-')
plt.show()
```
